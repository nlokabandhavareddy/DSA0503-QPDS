2.Write a Pandas program to display the IDs of employees who have done two or more jobs in the past.

```python
import pandas as pd
data = {
    'EMPLOYEE_ID': [102,101,101,201,114,122,200,176,176,200],
    'START_DATE': [
        '2001-01-13','1997-09-21','2001-10-28','2004-02-17','2006-03-24',
        '2007-01-01','1995-09-17','2006-03-24','2007-01-01','2002-07-01'
    ],
    'END_DATE': [
        '2006-07-24','2001-10-27','2005-03-15','2007-12-19','2007-12-31',
        '2007-12-31','2000-06-17','2006-12-31','2007-12-31','2006-12-31'
    ],
    'JOB_ID': [
        'IT_PROG','AC_ACCOUNT','AC_MGR','MK_REP','ST_CLERK',
        'ST_CLERK','AD_ASST','SA_REP','SA_MAN','AC_ACCOUNT'
    ],
    'DEPARTMENT_ID': [60,110,110,20,50,50,90,80,80,90]
}
df = pd.DataFrame(data)
result = df.groupby('EMPLOYEE_ID').size()
employees_with_multiple_jobs = result[result >= 2].index.tolist()
print("Employees with 2 or more jobs:")
print(employees_with_multiple_jobs)
```

4.Write a Pandas program to create a line plot of the historical stock prices of Alphabet Inc. between two specific dates.

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
symbol = "GOOGL"
start_date = "2022-01-01"
end_date = "2022-06-30"
df = yf.download(symbol, start=start_date, end=end_date)
print(df.head())
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['Close'])
plt.title("Alphabet Inc. (GOOGL) Stock Price")
plt.xlabel("Date")
plt.ylabel("Closing Price (USD)")
plt.grid(True)
plt.show()
```

5.Write a Pandas program to create a bar plot of the trading volume of Alphabet Inc. stock between two specific dates.

```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05'],
    'Volume': [1200000, 1500000, 1700000, 1600000, 1800000]
}
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
start = '2023-01-01'
end = '2023-01-05'
filtered = df.loc[start:end]
plt.bar(filtered.index, filtered['Volume'])
plt.xlabel("Date")
plt.ylabel("Trading Volume")
plt.title("Alphabet Inc. Trading Volume")
plt.show()
```

7.Write a Pandas program to create a Pivot table and find the maximum and minimum sale value of the items.(refer sales_data table)

```python
import pandas as pd
data = {
    'Item': ['Pen','Pen','Book','Book','Pencil','Pencil'],
    'Sale': [120, 150, 300, 280, 50, 40]
}
df = pd.DataFrame(data)
pivot = df.pivot_table(values='Sale', index='Item', aggfunc=['max','min'])
print(pivot)
```

8.Write a Pandas program to create a Pivot table and find the item wise unit sold. .(refer sales_data table)

```python
import pandas as pd
data = {
    'Item': ['Pen','Pen','Book','Book','Pencil','Pencil'],
    'Units': [10, 15, 20, 25, 5, 8]
}
df = pd.DataFrame(data)
pivot = df.pivot_table(values='Units', index='Item', aggfunc='sum')
print(pivot)
```

9.Write a Pandas program to create a Pivot table and find the total sale amount region wise, manager wise, sales man wise. .(refer sales_data table)

```python
import pandas as pd
data = {
    'Region': ['East','Central','Central','Central','West'],
    'Manager': ['Martha','Hermann','Hermann','Timothy','Timothy'],
    'SalesMan': ['Alexander','Shelli','Luis','David','Stephen'],
    'Sale_amt': [113810, 250000, 43128, 6075, 67088]
}
df = pd.DataFrame(data)
pivot = df.pivot_table(values='Sale_amt',
                       index=['Region','Manager','SalesMan'],
                       aggfunc='sum')
print(pivot)
```

10.Create a dataframe of ten rows, four columns with random values. Write a Pandas program to highlight the negative numbers red and positive numbers black.

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.randn(10, 4), columns=list('ABCD'))
def highlight(val):
    color = 'red' if val < 0 else 'black'
    return f'8color: {color}'
df.style.applymap(highlight)
```
