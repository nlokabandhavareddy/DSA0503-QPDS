21.Write a Pandas program to swap the cases of a specified character column in a given DataFrame.

```python
import pandas as pd
df = pd.DataFrame({
    'name': ['Alberto', 'FRANCO', 'gInO']
})
df['name'] = df['name'].str.swapcase()
print(df)
```

23.Write a Python program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title.

```python
import matplotlib.pyplot as plt
with open('test.txt', 'w') as f:
    f.write('1 2\n')
    f.write('2 4\n')
    f.write('3 1\n')
x = []
y = []
for line in open("test.txt"):
    a, b = line.split()
    x.append(float(a))
    y.append(float(b))
plt.plot(x, y)
plt.xlabel("x - axis")
plt.ylabel("y - axis")
plt.title("Sample graph!")
plt.show()
```

24.Write a Python program to draw line charts of the financial data of Alphabet Inc. between October 3, 2016 to October 7, 2016.

```python
import pandas as pd
import matplotlib.pyplot as plt
with open('fdata.csv', 'w') as f:
    f.write('Date,Open,High,Low,Close\n')
    f.write('10-03-16,774.25,776.065002,769.5,772.559998\n')
    f.write('10-04-16,776.030029,778.710022,772.890015,776.429993\n')
    f.write('10-05-16,779.309998,782.070007,775.650024,776.469971\n')
    f.write('10-06-16,779,780.47998,775.539978,776.859985\n')
    f.write('10-07-16,779.659973,779.659973,770.75,775.080017\n')
df = pd.read_csv("fdata.csv")
df['Date'] = pd.to_datetime(df['Date'], format='%m-%d-%y')
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Open'], label='Open')
plt.plot(df['Date'], df['High'], label='High')
plt.plot(df['Date'], df['Low'], label='Low')
plt.plot(df['Date'], df['Close'], label='Close')
plt.xlabel("Date")
plt.ylabel("Price")
plt.title("Alphabet Inc. Financial Data")
plt.legend()
plt.grid(True)
plt.show()
```

25.Write a Python program to plot two or more lines with legends, different widths and colors.

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
plt.plot(x, [2, 4, 6, 8], label='Line 1', linewidth=3, color='red')
plt.plot(x, [1, 3, 2, 5], label='Line 2', linewidth=2, color='blue')
plt.plot(x, [4, 3, 5, 7], label='Line 3', linewidth=1, color='green')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Lines")
plt.legend()
plt.show()
```

27.Write a Python programming to display a bar chart of the popularity of programming Languages.

29.Write a Python programming to display a bar chart of the popularity of programming Languages. Use different color for each bar.

```python
import matplotlib.pyplot as plt

langs = ['Java','Python','PHP','JavaScript','C#','C++']
pop = [22.2,17.6,8.8,8,7.7,6.7]
colors = ['red','blue','green','orange','purple','cyan']

plt.bar(langs, pop, color=colors)
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.show()
```

30.Write a Python program to create bar plot of scores by group and gender. Use multiple X values on the same chart for men and women.

```python
import matplotlib.pyplot as plt
import numpy as np

men = [22, 30, 35, 35, 26]
women = [25, 32, 30, 35, 29]

x = np.arange(len(men))
width = 0.35

plt.bar(x - width/2, men, width, label='Men')
plt.bar(x + width/2, women, width, label='Women')

plt.legend()
plt.show()
```
