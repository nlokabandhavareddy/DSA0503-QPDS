31.Write a Python program to create a stacked bar plot with error bars.
Note: Use bottom to stack the women?s bars on top of the men?s bars.

```python
import matplotlib.pyplot as plt
import numpy as np

men = [22, 30, 35, 35, 26]
women = [25, 32, 30, 35, 29]
men_sd = [4, 3, 4, 1, 5]
women_sd = [3, 5, 2, 3, 3]

x = np.arange(len(men))

plt.bar(x, men, yerr=men_sd)
plt.bar(x, women, bottom=men, yerr=women_sd)

plt.show()
```

34.Write a Python program to draw a scatter plot using random distributions to generate balls of different sizes.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.rand(50) * 500   # random ball sizes

plt.scatter(x, y, s=sizes)
plt.show()
```
